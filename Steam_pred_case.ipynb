```python
import matplotlib.pyplot as plt
import matplotlib
import random

def generate_unique_colors(length):
    """Generate a list of unique colors in hex format using Matplotlib."""
    if length > 256:
        raise ValueError("Length exceeds the maximum number of unique colors available in the colormap.")
    
    # Use a colormap to generate distinct colors
    colormap = plt.get_cmap('hsv', length+1)  # HSV colormap for distinct colors
    hex_colors = [matplotlib.colors.rgb2hex(colormap(i)[:3]) for i in range(length)]    
    return hex_colors 
```

```python
def calculate_mean(filter):
    data_comb = pd.DataFrame()
    for name in filter:
        data_frag = filter[name]['data']
        data_frag = data_frag.groupby('days_diff_end').mean()  # Group by 'days_diff_end' and calculate mean
        data_select = data_frag[['price']]  # Selecting the 'price' column after grouping
        data_select.columns = [name]  # Setting the column name to the current case name
        data_comb = pd.concat([data_comb, data_select], axis=1)  # Concatenating along columns
        data_comb = data_comb.sort_index(ascending=False)
    return {'mean':data_comb.mean(axis=1), 'std':data_comb.std(axis=1)}
```

```python
""" 
Creating fetures and labels for regression
y - next years return in %
X - labels as categorical variables + last year return in % + mean return in % + change of volume in % + last year change in %
X and y is for every year for every name
"""
def tags_transform(sub_tags, tags):
    return [tag in sub_tags for tag in tags]

def data_transform(data):
    df = data['data']
    years = data['years']
    sub_tags = data['tag']
    data_2 = []
    for y in range(years):
        interval = (df['days_diff_beg'] > y * 365) & (df['days_diff_beg'] <= (y + 1) * 365)
        if np.sum(interval) <2:
            data_2.append([None,None])
        else:
            
            data_2.append([np.median(df['price'][interval]), np.log(np.sum(df['volume'][interval]))])
    data_2 = pd.DataFrame(data_2, columns = ['price', 'volume'])
    return data_2
        
X_train = []
y_train = []
X_test = []
y_test = []
names_train = []
names_test = []
X_now = []
names_now = []
price_now = []
X_all = []
y_all = []
names_all = []
for name in data_formated:
# for name in {'Glove Case':data_formated['Glove Case']}:
    data_raw = data_formated[name]
    data_tags = tags_transform(data_raw['tag'], tags+['gloves', 'knife', 'regular'])
    data_pv = data_transform(data_raw)
    data_prim = (data_pv.iloc[1:,:].reset_index(drop=True) - data_pv.iloc[:-1,:].reset_index(drop=True)) / data_pv.iloc[:-1,:].reset_index(drop=True)
    if len(data_prim) > 1: 
        for i in range(1,len(data_prim)):
            d_mean = data_prim.iloc[:i, :].mean().values.tolist()
            d_last_vol =  data_prim.iloc[i-1, 1]
            d_last_price = data_prim.iloc[i-1, 0]
            d_y = data_prim.iloc[i, 0]
            if not any(np.isnan(value) for value in d_mean+[d_last_price, d_y]):
                names_train.append(name)
                names_all.append(name)
                # X_train.append(d_mean+[d_last_price, d_last_vol]+data_tags)
                X_train.append(d_mean+[d_last_price]+data_tags)
                X_all.append(d_mean+[d_last_price]+data_tags)
                # X_train.append(d_mean+data_tags)
                y_train.append(d_y)
                y_all.append(d_y)
        X_test.append(X_train.pop())
        y_test.append(y_train.pop())
        names_test.append(names_train.pop())

        d_mean = data_prim.mean().values.tolist()
        d_last_vol =  data_prim.iloc[-1, 1]
        d_last_price = data_prim.iloc[-1, 0]
        if not any(np.isnan(value) for value in d_mean+[d_last_price, d_y]):
            names_train.append(name)
            X_now.append(d_mean+[d_last_price]+data_tags)
            names_now.append(name)
            price_now.append(data_pv.iloc[-1, 0])
            
print(len(y_train))
print(len(y_test))
print(len(X_now))
print(len(y_all))
```
